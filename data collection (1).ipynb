```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
```

```python
saree_img_link = []
jeans_img_link = []
shorts_link = []
tshirts_link = []
```

```python
def product_images(url,product_link):
    
    product = input('Enter the name : ')
    driver = webdriver.Chrome(r'chromedriver.exe')
    driver.maximize_window() #maximizing browser windows
    driver.get(url)
    
    #search tab
    search_bar = driver.find_element_by_id('twotabsearchtextbox')
    search_bar.click()
    search_bar.send_keys(product)
    search_but = driver.find_element_by_id('nav-search-submit-text')
    search_but.click()
    
    for page in range(0,6):
        
        #image element
#         WebDriverWait(driver,2).until(EC.presence_of_all_elements_located((By.CLASS_NAME, 's-image')))
        images = driver.find_elements_by_xpath("//img[@class='s-image']")
        
        for i in images:
            img_link = i.get_attribute('src')
            if img_link is not None:
                if(img_link[0:5]=='https'):
                    product_link.append(img_link)
            
    
        pagination = driver.find_element_by_xpath("//ul[@class='a-pagination']")
        next_page = pagination.find_element_by_xpath(".//li[@class='a-last']")
        next_page.click()
        time.sleep(2)
```
